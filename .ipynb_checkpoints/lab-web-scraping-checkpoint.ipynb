```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from unidecode import unidecode  # Library to clean up special characters

# Rating conversion dictionary
rating_map = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}

# Function to clean text by removing special characters
def clean_text(text):
    return unidecode(text)  

# Scraping function
def scrape_books(min_rating=4, max_price=20):
    base_url = "http://books.toscrape.com/catalogue/"
    page_url = "http://books.toscrape.com/catalogue/page-{}.html"
    
    all_books = []
    
    for page in range(1, 51):  # 50 page in website
        response = requests.get(page_url.format(page))
        
        # Ensure correct decoding 
        response.encoding = 'utf-8'  # Force proper encoding to avoid 'Â' character issues
        
        soup = BeautifulSoup(response.text, "html.parser")
        
        # Get all books in the page
        books = soup.find_all(class_="product_pod")
        
        for book in books:
            title = clean_text(book.h3.a["title"])  # Clean title
            
            # Fix price extraction by removing £ and extra characters
            price_str = book.find(class_="price_color").text.strip()
            price_str = price_str.replace('£', '').replace('Â', '').replace(',', '').strip()  # Clean the string
            price = float(price_str)  # Convert to float
            
            rating = rating_map[book.p["class"][1]]  # Get rating from class names
            availability = clean_text(book.find(class_="availability").text.strip())  # Clean availability
            link = book.h3.a["href"]
            
            if price <= max_price and rating >= min_rating:
                book_url = base_url + link
                book_detail_response = requests.get(book_url)
                book_soup = BeautifulSoup(book_detail_response.text, "html.parser")
                
                # Get the product details table
                product_info = book_soup.find(class_="table-striped")
                upc = product_info.find_all("td")[0].text  # Extract UPC
                
                genre = clean_text(book_soup.find("ul", class_="breadcrumb").find_all("a")[2].text)  # Extract Genre
                
                # Get description if is available
                try:
                    description = book_soup.find(id="product_description").find_next_sibling("p").text
                    description = clean_text(description)  # Clean description
                except AttributeError:
                    description = "No description available"
                
                # Append book data to list
                all_books.append({
                    "UPC": upc,
                    "Title": title,
                    "Price (£)": price,
                    "Rating": rating,
                    "Genre": genre,
                    "Availability": availability,
                    "Description": description
                })
    
    # Convert the list of books to a DataFrame
    df = pd.DataFrame(all_books)
    return df
```

```python
min_rating = 4
max_price = 20
books_df = scrape_books(min_rating=min_rating, max_price=max_price)
    
# Display DataFrame
books_df.head(20)
```
